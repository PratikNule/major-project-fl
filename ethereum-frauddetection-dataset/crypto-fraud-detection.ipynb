## Introduction
This document describes a dataset for detecting fraudulent Ethereum accounts based on transaction patterns, ERC20 token activities, and account behaviors. The dataset is unbalanced, which should be taken into account during analysis and modeling.

## Objective
The goal is to predict fraudulent Ethereum transactions using historical transaction data. This is a binary classification problem where:

- **FLAG = 1** indicates fraud
- **FLAG = 0** indicates non-fraud

# Oversampling

```python
# Separate features (X) and target (y)
X = df.drop('FLAG', axis=1)  # Features
y = df['FLAG']  # Target (fraud flag)
```

```python
# Over-sample the minority class using SMOTE and clean with ENN (Edited Nearest Neighbors)
from imblearn.combine import SMOTEENN

# Initialize SMOTEENN
smote_enn = SMOTEENN(random_state=42)

# Resample the dataset
X_resampled, y_resampled = smote_enn.fit_resample(X, y)

# Check shapes after resampling
X_resampled.shape, y_resampled.shape
```

```python
import os
```

```python
# Step 3: Convert to DataFrame and add target column
global_train_df = pd.DataFrame(X_resampled, columns=X.columns)
global_train_df['FLAG'] = y_resampled

global_test_df = X.copy()
global_test_df['FLAG'] = y

# Step 4: Save to CSV
global_train_df.to_csv(os.path.join(OUTPUT_DATA_DIR, "global_train_centralized.csv"), index=False)
global_test_df.to_csv(os.path.join(OUTPUT_DATA_DIR, "global_test_centralized.csv"), index=False)

print("✅ Saved resampled global train set and original test set for centralized model.")
```

# Normalizing

```python
import joblib
```

```python
# Apply a power transform featurewise to make data more Gaussian-like
# Our features are not normally distributed and have many outliers

from sklearn.preprocessing import PowerTransformer

# Initialize the scaler
scaler = PowerTransformer()

# Fit and transform the training data 
X_train_transformed = scaler.fit_transform(X_train)

# Transform the test data (BUT DON'T FIT!)
X_test_transformed = scaler.transform(X_test)

# Save the scaler to a file using joblib
joblib.dump(scaler, os.path.join(OUTPUT_DATA_DIR, 'scaler.joblib'))
```

# Modeling

```python
# Generate logspace of 20 values for cross validation on 'C' hyperparameter
C_values = np.logspace(-2, 5, num=20)  # 20 values between 10^-2 and 10^5 
C_values
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# Initialize lists to store results
cv_scores = []

# Loop over C values
for C in C_values:
    # Initialize Logistic Regression with the current C value
    model = LogisticRegression(
        C=C,
        random_state=10,
        max_iter=5000  # Ensure convergence
    )
    
    # Perform 5-fold cross-validation
    scores = cross_val_score(
        estimator=model,
        X=X_train_transformed,  # Normalized training data
        y=y_train,        # Target variable
        scoring='roc_auc', # Use ROC-AUC for imbalanced data
        cv=5,             # 5-fold cross-validation
    )
    
    # Store the mean score for this C value
    cv_scores.append(scores.mean())

# Combine C values and their corresponding scores
cv_results = pd.DataFrame({
    'C': C_values,
    'ROC-AUC Score': cv_scores
})

# Display the results
print("Cross-Validation Results:")
print(cv_results)
```

```python
# Find the best C value
best_C = cv_results.loc[cv_results['ROC-AUC Score'].idxmax(), 'C']
best_score = cv_results['ROC-AUC Score'].max()

print(f"Best C value: {best_C}")
print(f"Best ROC-AUC Score: {best_score}")
```

```python
# Initialize the final model with the best C value
final_model = LogisticRegression(
    C=best_C,
    random_state=10,
    max_iter=5000
)

# Train the model on the full training set
final_model.fit(X_train_transformed, y_train)

# Predict on the test set
y_pred = final_model.predict(X_test_transformed)
y_proba = final_model.predict_proba(X_test_transformed)[:, 1]  # Probabilities for AUC
```

```python
from sklearn.metrics import classification_report, roc_auc_score

# Generate classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Calculate ROC-AUC
print(f"ROC-AUC Score: {roc_auc_score(y_test, y_proba):.3f}")
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='coolwarm')

# Add labels, title, and axis ticks
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')

# Display the plot
plt.show()
```

Given that we are working on fraud detection, we prioritize minimizing false negatives (132) over false positives (172), as missing a fraudulent transaction (false negative = TYPE II ERROR) is more costly than incorrectly flagging a legitimate transaction (false positive = TYPE I ERROR). In my opinion this is a good result for Logistic Regression, but let's try another model to see if we can further improve performance.

We have to limit TYPE II and TYPE I errors as much as possible, prioritizing on TYPE II.


# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest model
model = RandomForestClassifier(
    random_state=10,          # For reproducibility
    n_estimators=50,        # Number of trees, chose 50 for faster training and due to the small size of this dataset
)

# Train the model
model.fit(X_train_transformed, y_train)

# Predict on the test set
y_pred = model.predict(X_test_transformed)
y_proba = model.predict_proba(X_test_transformed)[:, 1]  # Probabilities for ROC-AUC
```

```python
# Generate classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Calculate ROC-AUC
print(f"ROC-AUC Score: {roc_auc_score(y_test, y_proba):.3f}")
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='coolwarm')

# Add labels, title, and axis ticks
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')

# Display the plot
plt.show()
```

We can observe that the results are significantly better than with Logistic Regression. Only 12 transactions were marked as false negatives, representing approximately 0.787% of fraudulent transactions.

We can further reduce this number through hyperparameter optimization to fine-tune the model and improve its performance.

RandomForest is perfect for this dataset because it:
